```python
numc=df.select_dtypes(include=['Int64','float']).columns
catc=df.select_dtypes(include=['object']).columns

```

```python
for n in numc:
    df[n]=df[n].fillna(df[n].mean())
   
```

```python
df.head()
```

```python
for c in catc:
    df[c]=df[c].fillna(df[c].mode()[0])
```

```python
df.shape
```

```python
df["Date of Review"]=pd.to_datetime(df["Date of Review"])
```

```python
df = df.drop_duplicates()
```

```python
df['Performance Score'] = (df['Performance Rating'] * 0.4) + (df['Tasks Completed'] * 0.3) + (df['Customer Feedback'] * 0.3)
```

```python
df.head()
```

```python
def increment(score):
    if score > 16:
        return 0.15
    elif 10 <= score <= 16:
        return 0.10
    else:
        return 0.05
    
```

```python
df["Updated_Salary"] = df["Salary"] * (1 + df["Salary_increment"])
```

```python
df["salary_increment"]=df["Performance Score"].apply(increment)
```

```python
df["Updated_Salary"] = df["Salary"] * (1 + df["salary_increment"])
```

```python
P_score_cutoff = df["Performance Score"].quantile(0.9)
salary_cutoff = df["Updated_Salary"].quantile(0.9)
```

```python
top_employees = df[
    (df["Performance Score"] >= P_score_cutoff) &
    (df["Updated_Salary"] >= salary_cutoff)
]
```
