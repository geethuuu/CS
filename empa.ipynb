```python
import pandas as pd
```

```python
df=pd.read_csv("employee_performance(in).csv")
```

```python
df
```

```python
print(df.to_string())
```

```python
df.isnull().sum()
```

```python
num_col=df.select_dtypes(include=['number']).columns
```

```python
cate_col=df.select_dtypes(include=['object','string']).columns
```

```python
df[num_col]=df[num_col].fillna(df[num_col].mean())
```

```python
df[num_col]
```

```python
for cate in cate_col:
    df[cate]=df[cate].fillna(df[cate].mode()[0])
```

```python
print(df)
```

```python
print(df.duplicated().to_string())
```

```python
df['Date of Review'] = pd.to_datetime(df['Date of Review'])
```

```python
df['Performance Score'] = (df['Performance Rating'] * 0.4) + (df['Tasks Completed'] * 0.3) + (df['Customer Feedback'] * 0.3)
```

```python
def get_increment(score):
    if score > 16:
        return 0.15
    elif score >= 10:
        return 0.10
    else:
        return 0.05

df['Salary Increment %'] = df['Performance Score'].apply(get_increment)
df['New Salary'] = df['Salary'] * (1 + df['Salary Increment %'])
```

```python
df
```

```python
dept_avg = (
    df
    .groupby('Department')
    .agg({
        'Performance Score': 'mean',
        'Salary': 'mean',
        'Working Hours': 'mean'
    })
    .reset_index()
)

dept_avg
```

```python
df['Month'] = df['Date of Review'].dt.to_period('M')
```

```python
monthly_perf = (
    df.groupby(['Department', 'Month'])
      [['Performance Score', 'Salary', 'Working Hours']]
      .mean()
      .reset_index()
)
```

```python
monthly_perf
```

```python
perf_cutoff = df['Performance Score'].quantile(0.90)
salary_cutoff = df['New Salary'].quantile(0.90)
```

```python
top_10_percent = df[
    (df['Performance Score'] >= perf_cutoff) &
    (df['New Salary'] >= salary_cutoff)
]
```

```python
top_10_percent.columns
```

```python
top_10_percent[['Employee Name', 'Department', 'Performance Score', 'New Salary']]
```
